```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Таблица Бучера:



```python
A=1
B=3
X_0 = np.ones(2)

def f(t, x):
    y = np.zeros(2)
    y[0] = A + x[0]**2 * x[1] - (B + 1) * x[1]
    y[1] = B * x[0] - x[0]**2 * x[1]
    
    return y
```

```python
def rk_explicit_step(x_prev, tau, t, f, a, c):
    k = [f(t+tau*c[0], x_prev)]
    for i in range(1, len(c)):
        cur_arg = x_prev
        for j in range(i):
            cur_arg += tau*a[i, j]*k[j]
        k.append(f(t+tau*c[i], cur_arg))
    return k

def solve_rk(f, x_0, time, N, butcher_coefs, rk_stepper=rk_explicit_step):
    '''
    :args:
        f - правая часть ДУ, f(t, x)
        x_0 - начальное условие
        time - промежуток времени, сетки
        N - число узлов в сетке
        butcher_coefs - словарь матрицы бутчера (см. картинку выше) 
    '''
    tau = (time[1] - time[0]) / N
    t = tau
    xs = [x_0]
    for i in range(1, N):
        t += tau
        k = rk_stepper(xs[-1], tau, t, f, butcher_coefs['a'], butcher_coefs['c'])
        xs.append(xs[-1] + tau*np.dot(butcher_coefs['b'], k))
    
    return np.array(xs)
```

```python
def plot_trajectory(x):
    plt.figure(figsize=(10,8))
    plt.plot(x[:, 0], x[:, 1])
    plt.xlabel('u')
    plt.ylabel('v')
    plt.show()
    pass
```

```python
print('1ый, порядок')
butcher1 = {
    'a' : np.array([
        [[0]]
    ]),
    'b' : np.array([1]),
    'c' : np.array([0])
}

answ_x = solve_rk(f, X_0, [0, 50], 10000, butcher1)
plot_trajectory(answ_x)
```
