```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from scipy import optimize
```

Таблица Бучера:



```python
X_0 = [2, 0]
A = 1e3    

def f(x):
    y = np.zeros(2)
    y[0] = x[1]
    y[1] = -1 * A * (x[1]*(x[0]**2-1)+x[0])
    
    return y
```

```python
def rk_implicit_step(x_prev, tau, t, f, a, c):
    '''
    будем на каждом шаге искать k_i при помощи метода ньютона - крылова для поиска корней нелинейной СУ
    '''
    
    def G(k):
        k = k.reshape((len(c), len(x_prev)))

        k_new = np.array([
            f(x_prev+tau*(np.dot(a[i, :], k))) for i in range(len(c))
        ]).flatten()
        k = k.flatten()

        k = k - k_new
        return k

    try: 
        k =  optimize.newton_krylov(
            G, np.zeros(len(c) * len(x_prev))
        ).reshape((len(c), len(x_prev)))
    except Exception as e:
        if isinstance(e.args[0], str):
            k = np.zeros((len(c), len(x_prev)))
        else:
            k = e.args[0].reshape((len(c), len(x_prev)))

    return k

def solve_rk(f, x_0, time, N, butcher_coefs, rk_stepper=rk_implicit_step):
    '''
    :args:
        f - правая часть ДУ, f(t, x)
        x_0 - начальное условие
        time - промежуток времени, сетки
        N - число узлов в сетке
        butcher_coefs - словарь матрицы бутчера (см. картинку выше) 
    '''
    tau = (time[1] - time[0]) / N
    t = tau
    xs = [x_0]
    for i in tqdm(range(1, N)):
        t += tau
        k = rk_stepper(xs[-1], tau, t, f, butcher_coefs['a'], butcher_coefs['c'])
        xs.append(xs[-1] + tau*np.dot(butcher_coefs['b'], k))
    
    return np.array(xs)
```

```python
def plot_trajectory(x):
    plt.figure(figsize=(10,8))
    plt.plot(x[:, 0], x[:, 1])
    plt.xlabel('y1')
    plt.ylabel('y2')
    plt.show()
    pass
```

```python
butcher1 = {
    'a' : np.array([
        [5/12, -1/12],
        [3/4, 1/4]
    ]),
    'b' : np.array([3/4, 1/4]),
    'c' : np.array([1/3, 1])
}

butcher2 = {
    'a' : np.array([
        [0, 0],
        [1/2, 1/2]
    ]),
    'b' : np.array([1/2, 1/2]),
    'c' : np.array([0, 1])
}

butcher3 = {
    'a' : np.array([
        [1/4, 1/4 - 1/6*np.sqrt(3)],
        [1/4 - 1/6*np.sqrt(3), 1/4]
    ]),
    'b' : np.array([1/2, 1/2]),
    'c' : np.array([1/2-1/6*np.sqrt(3), 1/2+1/6*np.sqrt(3)])
}
```

```python
answ_x = solve_rk(f, X_0, [0, 20], 100, butcher1)
```

```python
plot_trajectory(answ_x)
```
