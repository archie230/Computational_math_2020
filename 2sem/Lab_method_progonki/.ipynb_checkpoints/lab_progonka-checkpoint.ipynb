```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values = np.zeros(10)
N = 10
e = 0.0001
a = 0.0
b = 1.0

def k(x):
    return np.exp(np.sin(x))

def q(x):
    return np.exp(np.cos(x))

def f(x):
    return 1
```

```python
def Matrix(N, h):
    L = np.zeros(N * N).reshape(N, N)
    d = np.zeros(N)

    L[0][0] = -1.0
    L[0][1] = 1.0
    L[N - 1][N - 2] = -1.0 * k(1.0)
    L[N - 1][N - 1] = h + k(1.0)
    for i in range(1, N - 1):
        L[i][i - 1] = k((float(i) - 0.5) * h) / h / h
        L[i][i] = -1.0 * (k((float(i) - 0.5) * h) + k((float(i) + 0.5) * h)) / h / h - q(i * h)
        L[i][i + 1] = k((float(i) + 0.5) * h) / h / h
        d[i] = -1.0 * f(i * h)
    return L, d
```

```python
def additional(L, d):
    N = d.shape[0]
    p = np.zeros(N)
    r = np.zeros(N)

    p[0] = L[0][1] / L[0][0]
    r[0] = d[0] / L[0][0]

    for i in range(1, N - 1):
        p[i] = L[i][i + 1] / (L[i][i] - L[i][i - 1] * p[i - 1])
        r[i] = (d[i] - L[i][i - 1] * r[i - 1]) / (L[i][i] - L[i][i - 1] * p[i - 1])
    r[N - 1] = (d[N - 1] - L[N - 1][N - 2] * r[N - 2]) / (L[N - 1][N - 1] - L[N - 1][N - 2] * p[N - 2])

    return p, r
```

```python
fig = plt.figure(figsize = [20,20])
ax = fig.subplots()
N = 10
print(f'Решение для {N} точек в сетке')
while 1:
    h = (b - a) / float(N - 1)
    L, d = Matrix(N, h)
    p, r = additional(L, d)

    u = np.zeros(N)
    u[N - 1] = r[N - 1]
    
    for i in range(N - 2, -1, -1):
        u[i] = r[i] - p[i] * u[i + 1]
    x = np.array([h * t for t in range(N)], dtype=float)
     
    ax.plot(x, u)
    
    if all([abs(values[i] - u[int(1/10.0) * i]) < e for i in range(10)]):
        ax.set_xlabel("x", fontsize=20)
        ax.set_ylabel("u", fontsize=20)
        ax.grid()
        break
    else:
        values = np.array([u[int(1/11.0) * i] for i in range(11)])
    N *= 2
    
```
