```python
y0 = 1.
y1 = 0.
h = 1e-4  # ширина шага для рк и для метода ньютона
eps = 1e-3
X = 1 # правая граница
```

```python
BUTCHER = {
    'a' : np.array([ 
            np.array( [0.,  0.]),
            np.array( [0.5, 0.]) 
        ]),
    'b' : np.array( [0., 1.])
}
```

```python
def method_rk(alpha):
    x_i = 0
    y_i = np.array([1., alpha])
    x = []
    y = []

    while x_i < X:
        x.append( x_i)
        x_i += h
        y.append( y_i)
        y_i = step(y_i, h)
    return(x, np.transpose(np.array(y)))

def step(y_n, h):
    K = np.zeros((2, len(BUTCHER['b'])), dtype=float)
    K[0] = func(y_n)
    for i in range (1, len(BUTCHER['b'])):
        K[i] = func(y_n + h * BUTCHER['a'][i][i-1] * K[i-1])

    return y_n + h * np.array([sum([BUTCHER['b'][j] * K[j][i] for j in range(len(BUTCHER['b']))]) for i in range(len(y_n))])

def func(y):
    return np.array([y[1], (y[1] ** 2)])
```

```python
def alpha():
    alpha1 = -1.7
    alpha2 = 0.
    while(abs(alpha2 - alpha1) > eps):
        alpha2 = alpha1
        alpha1 = alpha2 - (F(alpha2) - y1) / dF(alpha2)
    return alpha1

def F(alpha):
    return (method_rk(alpha))[1][0][-1]

def dF(alpha):
    return ((method_rk(alpha + h))[1][0][-1] - (method_rk(alpha)[1][0][-1])) / h
```

```python
alpha = alpha()
print(f'alpha : {alpha}')

plt.figure(figsize=(10, 8))
res = method_rk(alpha)
plt.plot(res[0], res[1][0])
plt.show()
```
