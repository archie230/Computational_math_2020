### Инициализация матрицы из 10.6 (д)

```python
A = np.zeros((100, 100))
f = np.arange(1, 101)

ones = 5

for i in range(A.shape[0]):
    for j in range(ones):
        if i == j:
            A[i, j] = 10
        else:
            A[i, j] = 1
            
    if ones < A.shape[0]:
        ones += 1
        
A
```

### Алгоритм Гаусса

```python
def FirstNonZero(A, index):
    column = index
    
    while(A[index, column] != 0 and column < A.shape[0]):
        column += 1
    
    if column == A.shape[0]:
        return -1
    else:
        return column

# Gauss algorithm
def SolveGauss(A, f):
    assert(A.ndim == 2)
    assert(A.shape[0] == A.shape[1])
    
    A = np.hstack((A, f[:, np.newaxis]))
    
    # straight run
    for i in range(A.shape[0]):
        if A[i, i] == 0:
            non_zero_idx = FirstNonZero(A, i)
            if non_zero_idx < 0:
                continue
            else:
                A[[i, non_zero_idx]] = A[[non_zero_idx, i]]
        
        for j in range(i + 1, A.shape[0]):
            A[j, :] -= (A[j, i] / A[i, i]) * A[i, :]
    
    X = np.zeros(A.shape[0])
    
    # reverse run
    X[-1] = A[-1, -1] / A[-1, -2]
    for i in range(A.shape[0] - 2, -1, -1):
        X[i] = (A[i, -1] - A[i, i+1:-1].dot(X[i+1:])) / A[i, i]
    
    return X
```

```python
#smoke test
X=SolveGauss(A, f)
assert(np.allclose(A.dot(X), f)), "Oh oh smth went wrong"
```

### Итерационный метод зейделя

**Возьмем критерий останова :** $ \|u_{k+1}-u_{k}\| \lt \varepsilon $

```python
def Zeidel(u_0, A, f, eps=1e-2):
    # initializing matrices
    L = np.zeros((A.shape[0], A.shape[1]))
    D = np.zeros((A.shape[0], A.shape[1]))
    U = np.zeros((A.shape[0], A.shape[1]))

    for i in range(L.shape[0]):
        for j in range(L.shape[1]):
            if (j < i):
                L[i, j] = A[i, j]

    for i in range(L.shape[0]):
        for j in range(L.shape[1]):
            if (j == i):
                D[i, j] = A[i, j]

    U = A - D - L
    B = -np.linalg.inv(L + D).dot(U)
    F = np.linalg.inv(L + D).dot(f)
    
    # starting iteration
    u_k = u_0
    i = 0
    
    while True:
        i += 1
        u_kplus = B.dot(u_k) + F
        
        print("Невязка на {} шаге: {} ".format(i, A.dot(u_kplus) - f))
        print("Её норма: ", np.linalg.norm(A.dot(u_kplus) - f), "\n")
        
        if np.linalg.norm(u_kplus - u_k) < eps:
            return u_kplus
        u_k = u_kplus
    
```

```python
answ = Zeidel(np.zeros(A.shape[0]), A, f, eps=1e-5)

assert(np.allclose(A.dot(answ), f)), "Oh oh smth went wrong"
```
