```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 6))

x = np.linspace(1, 20, 1000)
plt.plot(x, np.log(x) / x**2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\frac{lnx}{x^{2}}$")
plt.show()
```

Найдем значение функции на полувысоте:

$
\frac{df}{dx} = 0 \cap x \geq 1
$

$
\frac{x-2xlnx}{x^4} = 0 \cap x \geq 1
$

$
x_{max} = \sqrt{e}
$

$
f(x_{max}) = \frac{1}{2e}
$

$
\frac{f(x_{max})}{2} = \frac{1}{4e}
$

Будем решать МПИ и Методом Ньютона уравнение: $\frac{lnx}{x^2} = \frac{1}{4e} \cap x \geq 1$

$x_1, x_2$ - решения, тогда $|x_2 - x_1|$ - искомая ширина функции

### Метод Простой Итерации

```python
def SimpleIteration(x_0, F, eps=1e-3, Nmax=10000):
    x_k = x_0
    x_kplus1 = F(x_k)
    N = 1
    
    while np.abs(x_k - x_kplus1) > eps and N < Nmax:
        x_k = x_kplus1
        x_kplus1 = F(x_k)
        N += 1
    
    if (N == Nmax):
        print("Max iteration number reached!, $x_0$: {} , $x_{Nmax}$: {}".format(x_0, x_kplus1))
        
    return x_kplus1, N
```

#### Для $x_2$ : 

$x = 4e \frac{lnx}{x} = F(x)$

т.к.
$F^{'}(x) = 4e \frac{1-lnx}{x^2}$, $|F^{'}(x)| < 1 \cap x \geq 2.5$, что видно из графика ниже

поэтому $F(x)$ - сжимающее отображение на $x \geq 2.5$ и $x_{k+1} = F(x_{k})$ сходится на $x \geq 2.5$

```python
plt.figure(figsize=(10, 6))

plt.plot(x, 4 * np.exp(1) * (1-np.log(x)) / x**2, label="$F^{'}(x)=4e\\frac{1-lnx}{x^{2}}$")
plt.hlines(1, 0, 20, label="$y=1$")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("$4e\\frac{1-lnx}{x^{2}}$")
plt.show()
```

```python
F = lambda x: 4*np.exp(1)*np.log(x)/x
x_2, N = SimpleIteration(4, F)

print("x_2: {}, iteration number: {}".format(x_2, N))
```

#### Для $x_1$ : 

$x = exp(\frac{x^2}{4e}) = F(x)$

т.к.
$F^{'}(x) = \frac{x}{2e} exp(\frac{x^2}{4e})$, $|F^{'}(x)| < 1 \cap x \in [1, 2.5]$, что видно из графика ниже

поэтому $F(x)$ - сжимающее отображение на $x \in [1, 2.5]$ и $x_{k+1} = F(x_{k})$ сходится на $x \in [1, 2.5]$

```python
plt.figure(figsize=(10, 6))

plt.plot(x[x<2.5], (x/(2*np.exp(1))*np.exp(x**2/(4*np.exp(1))))[x<2.5], label="$\\frac{x}{2e} exp(\\frac{x^2}{4e})$")
plt.hlines(1, 1, 2.5, label="$y=1$")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\frac{x}{2e} exp(\\frac{x^2}{4e})$")
plt.show()
```

```python
F = lambda x: np.exp(x**2/(4*np.exp(1)))
x_1, N = SimpleIteration(1.1, F)

print("x_1: {}, iteration number: {}".format(x_1, N))
```

```python
print("Итого получаем ширину функции на fmax/2: ", np.round(x_2-x_1, 3))
```

### Метод Ньютона

найти решения $$f(x) = \frac{lnx}{x^2} - \frac{1}{4e} = 0 \cap x \geq 1$$

$x_{k+1} = x_k - f_x^{-1}(x_k)f(x_k)$

$f^{'}(x) = \frac{1-2lnx}{x^3}$

$f^{''}(x) = \frac{6lnx-5}{x^4}$

```python
def NewtonMethod(x_0, f, df_inv, eps=1e-3, Nmax=10000):
    x_k = x_0
    x_kplus1 = x_k - df_inv(x_k) * f(x_k)
    N = 1
    
    while np.abs(x_k - x_kplus1) > eps and N < Nmax:
        x_k = x_kplus1
        x_kplus1 = x_k - df_inv(x_k) * f(x_k)
        N += 1
    
    if (N == Nmax):
        print("Max iteration number reached!, $x_0$: {} , $x_{Nmax}$: {}".format(x_0, x_kplus1))
        
    return x_kplus1, N
```

Исследуем $f^{'}$ и $f{''}$

```python
plt.figure(figsize=(10, 6))

plt.plot(x, (1-2*np.log(x))/x**3, label="$f^{'}$")
plt.plot(x, (6*np.log(x)-5)/x**4, label="$f^{''}$")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Newton Method")
plt.show()
```

#### Будем искать $x_2$ на $[2.5, 6]$

$|\frac{1}{f^{'}}| \leq 84$ and $|f^{''}(x)| \leq 1$ 

$x \in [2.5, 6]$

```python
plt.figure(figsize=(10, 6))

plt.plot(x[(x>2.5) & (x < 6)], (x**3/(1-2*np.log(x)))[(x>2.5) & (x<6)], label="$\\frac{1}{f^{'}}$")
plt.plot(x[(x>2.5) & (x < 6)], ((6*np.log(x)-5)/x**4)[(x>2.5) & (x < 6)], label="$f^{''}$")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Newton Method $\\frac{1}{f^{'}}$ and $f^{''}$")
plt.show()
```

Поэтому выберем $x_0$ так, чтобы $|f(x_0)| \leq \frac{q}{84^2} < \frac{1}{84^2}$ and $x_0 \in [2.5, 6]$

$f(x) = \frac{lnx}{x^2} - \frac{1}{4e}$

```python
x[((np.log(x)/x**2 - 1/(4*np.exp(1))) < 1/(84**2)) & (x>2.5) & (x<6)]
```

возьмем $x_0 = 4$

```python
f = lambda x: np.log(x)/x**2 - 1/(4*np.exp(1))
df_inv = lambda x : x**3/(1-2*np.log(x))

x_2, N = NewtonMethod(4, f, df_inv)

print("x_2: {}, iteration number: {}".format(x_2, N))
```

#### Будем искать $x_1$ на $[1, 1.6]$

$|\frac{1}{f^{'}}| \leq 69$ and $|f^{''}(x)| \leq 5$ 

$x \in [1, 1.6]$

```python
plt.figure(figsize=(10, 6))

plt.plot(x[(x>1) & (x < 1.6)], (x**3/(1-2*np.log(x)))[(x>1) & (x<1.6)], label="$\\frac{1}{f^{'}}$")
plt.plot(x[(x>1) & (x < 1.6)], ((6*np.log(x)-5)/x**4)[(x>1) & (x<1.6)], label="$f^{''}$")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Newton Method $\\frac{1}{f^{'}}$ and $f^{''}$")
plt.show()
```

Поэтому выберем $x_0$ так, чтобы $|f(x_0)| \leq \frac{q}{39^2 \cdot 5} < \frac{1}{39^2 \cdot 5}$ and $x_0 \in [1, 1.6]$

$f(x) = \frac{lnx}{x^2} - \frac{1}{4e}$

```python
x[((np.log(x)/x**2 - 1/(4*np.exp(1))) < 1/(39**2*5)) & (x>1) & (x<1.6)]
```

возьмем $x_0 = 1.1$

```python
f = lambda x: np.log(x)/x**2 - 1/(4*np.exp(1))
df_inv = lambda x : x**3/(1-2*np.log(x))

x_1, N = NewtonMethod(1.1, f, df_inv)

print("x_1: {}, iteration number: {}".format(x_1, N))
```

```python
print("Итого получаем ширину функции на fmax/2: ", np.round(x_2-x_1, 3))
```

### Вывод: 

- рассчитали методом Ньютона и МПИ ширину $\frac{lnx}{x^2}$ на полувысоте, она равна 2.693
- получилось убедиться, что метод Ньютона сходится быстрее МПИ, что соотвю теории, тк 
    - Метод ньютона $\sim \frac{1}{k^2} \: k \rightarrow \infty$
    - МПИ $\sim \frac{1}{k} \: k \rightarrow \infty$
